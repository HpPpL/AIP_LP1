```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
os.chdir("..")
os.chdir("Results")
LoF = os.listdir()
os.chdir("..")
# os.mkdir("Results_Graphs")

os.getcwd()
```

```python
for i in range(len(LoF)):
    os.chdir("Results")
    df = pd.read_table(LoF[i], skiprows=1, sep = ";")
    df.pop(' Message')
    os.chdir("..")

    arr_size = [1e+5, 2e+5, 3e+5, 4e+5, 5e+5, 6e+5, 7e+5, 8e+5, 9e+5, 1e+6]
    linestyles = list(mpl.lines.Line2D.lineStyles)[:3]

    if LoF[i][0:4] == "popB":
        name_of_plot = LoF[i][10:LoF[i].find(".")] + "popB time."

        arr_vec = []
        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_vec.append(df.iloc[i][0])
            arr_deque.append(df.iloc[i+10][0])
            arr_list.append(df.iloc[i+20][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_vec,  linestyles[0], label = 'Vector',)
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")

    elif LoF[i][0:4] == "popF":
        name_of_plot = LoF[i][10:LoF[i].find(".")] + "popF time."

        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_deque.append(df.iloc[i][0])
            arr_list.append(df.iloc[i+10][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")

    elif LoF[i][0:5] == "pushB":
        name_of_plot = LoF[i][11:LoF[i].find(".")] + "pushB time."

        arr_vec = []
        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_vec.append(df.iloc[i][0])
            arr_deque.append(df.iloc[i+10][0])
            arr_list.append(df.iloc[i+20][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_vec,  linestyles[0], label = 'Vector',)
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")

    elif LoF[i][0:5] == "pushF":
        name_of_plot = LoF[i][11:LoF[i].find(".")] + "pushF time."

        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_deque.append(df.iloc[i][0])
            arr_list.append(df.iloc[i+10][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")

    elif LoF[i][0:4] == "find":
        name_of_plot = LoF[i][10:LoF[i].find(".")] + " (find time)."

        arr_vec = []
        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_vec.append(df.iloc[i][0])
            arr_deque.append(df.iloc[i+10][0])
            arr_list.append(df.iloc[i+20][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_vec,  linestyles[0], label = 'Vector',)
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")

    elif LoF[i][0:4] == "sort":
        name_of_plot = LoF[i][10:LoF[i].find(".")] + "Sort time."

        arr_vec = []
        arr_deque = []
        arr_list = []

        for i in range(10):
            arr_vec.append(df.iloc[i][0])
            arr_deque.append(df.iloc[i+10][0])
            arr_list.append(df.iloc[i+20][0])


        fig, ax = plt.subplots()
        ax.plot(arr_size, arr_vec,  linestyles[0], label = 'Vector',)
        ax.plot(arr_size, arr_deque, linestyles[1], label = 'Deque')
        ax.plot(arr_size, arr_list, linestyles[2], label = 'List')

        ax.set_title(name_of_plot)
        ax.grid(axis='x', ls='--')
        ax.grid(axis='y', ls='--')
        ax.set_xlabel("Size of file")
        ax.set_ylabel("Time (ns)")

        ax.legend()

        os.chdir("Results_Graphs")
        plt.savefig(name_of_plot, dpi=300)
        os.chdir("..")
```
